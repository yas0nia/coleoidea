```python
#if there are repeating rows, remove them

df_drop = df.drop_duplicates(subset=['sseqid'], keep='first')
```

```python
df_drop
```

```python
#replace all semicolons with commas in column SMLF

df_drop['SMLF'] = df_drop['SMLF'].str.replace(';', ',')
```

```python
#for each row, check comma separated elements in column SMLF, if there are duplicates among these elements, remove these duplicates

df_drop['SMLF'] = df_drop['SMLF'].str.split(',')
df_drop['SMLF'] = df_drop['SMLF'].apply(lambda x: list(dict.fromkeys(x)))
```

```python
#remove brackets from column SMLF

df_drop['SMLF'] = df_drop['SMLF'].str.join(',')
```

```python
df_drop
```

```python
# for each row, calculate the number of editing sites (comma separated elements in a smlf column)
df_drop['num_changes'] = df_drop['SMLF'].apply(lambda x: len(x.split(',')))
```

```python
#for each comma separated element in column SMLF, check if first letter equals last letter, if so, remove it

def remove_syn(x):
    x = x.split(',')
    for i in x:
        if i[0] == i[-1]:
            x.remove(i)
    return ','.join(x)

df_drop['SMLF'] = df_drop['SMLF'].apply(remove_syn)
```

```python
df_drop['num_changes_nonsyn'] = df_drop['SMLF'].apply(lambda x: len(x.split(',')))
```

```python
import random

aminoacids = ['A', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'K', 'L', 'M', 'N', 'P', 'Q', 'R', 'S', 'T', 'V', 'W', 'Y']

#for ach row, randomly generate num_changes number of editing sites in format: aminoacid1 position (max is the len of 'Sequence') aminoacid2. 
# Aminoacid1 should be different from aminoacid2. 

def randomise(row):
    num_changes = row['num_changes']
    sequence = row['Sequence']
    smlf = row['SMLF']
    changes = []
    for i in range(num_changes):
        aminoacid1 = random.choice(aminoacids)
        position = random.randint(1, len(sequence)-1)
        aminoacid2 = random.choice(aminoacids)
        #while aminoacid2 == aminoacid1:
        #    aminoacid2 = random.choice(aminoacids)
        changes.append(aminoacid1 + str(position) + aminoacid2)
    return ','.join(changes)

df_drop['randomised'] = df_drop.apply(randomise, axis=1)
#df2['randomised'] = df2.apply(randomise, axis=1)
```
